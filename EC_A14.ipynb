### Ejercicio 1

### Ejercicio 3

````python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Punto(x={self.x},y={self.y})"
    
    def __eq__(self, coord):
        return self.x == coord.x and self.y == coord.y

    def __add__(self, other):
        if isinstance(other, int):
            nx = self.x + other
            return Punto(nx, self.y)
        
        if isinstance(other, Punto):
            nx = self.x + other.x
            ny = self.y + other.y
            return Punto(nx, ny)
        
        else:
            print("No puedo sumar esto")
            return None

pa = Punto(1, 2)
presult = Punto(4, 2)
pb = Punto(4, 5)

ph = Punto(5, 7)

assert(pa + 3 == presult)
assert(pa + pb == ph)
````

### Ejercicio 10

Una fracción también se puede multiplicar por otra fracción o por otro número entero.<br>
- La multiplicación de dos fracciones se realiza multiplicando sus numeradores y denominadores.<br>
- La multiplicación de una fracción por un número entero se realiza multiplicando el numerador por el número entero.<br>

Aplica los siguientes tests para comprobar la funcionalidad:<br>
```python
fa = Fraccion(3, 2)
fb = Fraccion(5, 3)

fc = Fraccion(15,6)
fd = Fraccion(9,2)

assert(fa * fb == fc)  # Debe devolver True
assert(fa * 3 == fd)  # Debe devolver True
assert(fa * 5 != fd)  # Debe devolver True
```


````python
class Fraccion:
    def __init__(self, numerador, denominador):
        self.numerador = numerador
        self.denominador = denominador
    
    def __eq__(self, other):
        return self.numerador == other.numerador and self.denominador == other.denominador
    
    def __ne__(self, other):
        return self.numerador != other.numerador or self.denominador != other.denominador
    
    def __mul__(self, other):

        if isinstance(other, int):
            nnum = self.numerador * other
            return Fraccion(nnum, self.denominador)
    
        if isinstance(other, Fraccion):
            nnum = self.numerador * other.numerador
            nden = self.denominador * other.denominador
            return Fraccion(nnum, nden)

fa = Fraccion(3, 2)
fb = Fraccion(5, 3)

fc = Fraccion(15,6)
fd = Fraccion(9,2)

assert(fa * fb == fc)
assert(fa * 3 == fd)
assert(fa * 5 != fd)
````

También se requiere que su representación (__repr__) sea `(Juan, García, 20, 1234)`.<br>

> 💡 Recuerda. El método __str__ se dispara a imprimir un objeto (print()) o también se puede llamar directamente al método __str__.<br>

Aplica los siguientes tests para comprobar la funcionalidad:<br>
```python
ea = Estudiante('Juan', 'García', 20, 1234)

### Ejercicio 12

Un estudiante está matriculado está matriculado en una serie de cursos.

- Crear una clase `Curso` con los atributos `nombre`.<br>
- Se requiere que un estudiante se pueda matricular en un curso o varios cursos.<br>
- Para ello crear un método `matricular` y pasar la el curso en el cual se quiere matricular.<br>
- Un estudiante tendrá una lista de cursos en los que está matriculado.<br>


Se requiere comprobar si un estudiante está matriculado en un curso, con el operador `in`, lo que hace necesario implementar el método `__contains__`.<br>

Aplica los siguientes tests para comprobar la funcionalidad:<br>
```python
ea = Estudiante('Juan', 'García', 20, 1234)
eb = Estudiante('Miguel', 'García', 20, 1234)

c1 = Curso('Matemáticas')
c2 = Curso('Física')
c3 = Curso('Química')

ea.matricular(c1)
ea.matricular(c2)

# Comparar si el estudiante `ea` está matriculado en el curso `c1`
assert(c1 in ea)  # Debe devolver True
assert(c2 in ea)  # Debe devolver True
assert(c3 not in ea)  # Debe devolver True

````python
class Estudiante:
    def __init__(self, nombre, apellidos, edad, numero_matricula):
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = edad
        self.numero_matricula = numero_matricula
        self.cursos_matriculados = []  

    def __str__(self):
        return f"Estudiante(nombre='{self.nombre}', apellidos='{self.apellidos}', edad={self.edad}, numero_matricula={self.numero_matricula})"

    def __repr__(self):
        return f"({self.nombre}, {self.apellidos}, {self.edad}, {self.numero_matricula})"
    
    def matricular(self, curso):
        self.cursos_matriculados.append(curso)

    def __contains__(self, curso):
        return curso in self.cursos_matriculados

class Curso:
    def __init__(self, nombre):
        self.nombre = nombre


ea = Estudiante('Juan', 'García', 20, 1234)
eb = Estudiante('Miguel', 'García', 20, 1234)

c1 = Curso('Matemáticas')
c2 = Curso('Física')
c3 = Curso('Química')

ea.matricular(c1)
ea.matricular(c2)

# Comparar si el estudiante `ea` está matriculado en el curso `c1`
assert c1 in ea  # Debe devolver True
assert c2 in ea  # Debe devolver True
assert c3 not in ea  # Debe devolver True

````
