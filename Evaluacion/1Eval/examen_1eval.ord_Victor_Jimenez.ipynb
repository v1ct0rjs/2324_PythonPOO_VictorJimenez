Crea una clase `Libro` que modele la información que se mantiene en una biblioteca sobre cada libro.

Atributos:
- Tiene que tener los atributos `titulo`, `autor`, `publicacion` (año) y `prestado` (Si/No).
- Un libro se crea con todos los atributos excepto `prestado`, que se inicializa a `False`.
- Todos los atributos son privados, y deben tener métodos `get` y `set` para acceder a ellos. (No @property)
  

Métodos:
- `prestar()`: Este método cambia el estado de `prestado` a `True`.
- `devolver()`: Este método cambia el estado de `prestado` a `False`.
- `info()`: Este método devuelve una cadena de texto que contiene toda la información del libro.

Para comprobar que la clase funciona correctamente, ejecuta el código disponible en la celda para tal efecto

````python

class Libro:
    def __init__(self, titulo, autor, publicacion, prestado = False):
        self._titulo = titulo
        self._autor = autor
        self._publicacion = publicacion
        self._prestado = prestado

    def info(self):
        return f'{self._titulo}, {self._autor}, {self._publicacion}'
    
    def get_prestado(self):
        return self._prestado
    
    def get_titilo(self):
        return self._titulo
    
    def get_autor(self):
        return self._autor
    
    def prestar(self):
        self._prestado = True
    
    def devolver(self):
        self._prestado = False
````

````python
# CELDA PARA TESTEAR TU CODIGO. Desomentar para usar.

libro1 = Libro("El Quijote", "Cervantes", 1605)
libro2 = Libro("El Hobbit", "Tolkien", 1937)

assert(libro1.info() == "El Quijote, Cervantes, 1605")
assert(libro2.get_prestado() == False)

libro1.prestar()
assert(libro1.get_prestado() == True)

libro1.devolver()
assert(libro1.get_prestado() == False)

libro1.get_autor()
````

### Ejercicio 3  (2,5 puntos)

Basado en el ejercicio anterior (copia todo el código de la clase `Libro`).

Agregar siguiente atributo:
- `paginas`: Número de páginas del libro (privado). Agrega al constructor, y se puedan obtener el valor de las págínas desde fuera de la clase. (No @property)

Agrega los siguientes métodos:

- Agrega a la clase `Libro` lo necesario para que se puedan comparar 2 libros con el operador `==` y `!=`. Dos libros son iguales si tienen el mismo título, autor.
- Agrega a la clase `Libro` para que se puedan sumar 2 libros:
  - Libro1 + Libro2: Devuelve un nuevo libro con las páginas de los dos libros sumadas.
  - Libro1 + 100: Devuelve un nuevo libro con las páginas del libro1 más 100.
  - Si se intenta sumar un libro con otro objeto que no sea un libro o ún número, debe retornar `None`.

Para comprobar que la clase funciona correctamente, ejecuta el código disponible en la celda para tal efecto

````python
class Libro:
    def __init__(self, titulo, autor, publicacion, paginas, prestado=False):
        self._titulo = titulo
        self._autor = autor
        self._publicacion = publicacion
        self._paginas = paginas
        self._prestado = prestado

    def info(self):
        return f'{self._titulo}, {self._autor}, {self._publicacion}, {self._paginas} páginas'

    def get_prestado(self):
        return self._prestado

    def get_titulo(self):
        return self._titulo

    def get_autor(self):
        return self._autor

    def get_paginas(self):
        return self._paginas

    def prestar(self):
        self._prestado = True

    def devolver(self):
        self._prestado = False

    def __eq__(self, other):
        return (self._titulo == other._titulo) and (self._autor == other._autor)

    def __add__(self, other):
        if isinstance(other, Libro):
            sp = self._paginas + other._paginas
            return Libro(self._titulo, self._autor, self._publicacion, sp, self._prestado)

        elif isinstance(other, int):
            sp = self._paginas + other
            return Libro(self._titulo, self._autor, self._publicacion, sp, self._prestado)

        else:
            return None
````

````python
# CELDA PARA TESTEAR TU CODIGO. Desomentar para usar.

libro1 = Libro("El hobbit", "Tolkien", 1937, 300)
libro2 = Libro("Harry Potter I", "J.K Rowling", 1997, 500)

assert((libro1 == libro2) is False) # False
assert(libro1 != libro2) # True

libro3 = libro1 + libro2
assert(libro3.get_paginas() == 800)

libro4 = libro1 + 100
assert(libro4.get_paginas() == 400)
````

La clase base `Vehiculo` se incluye en la celda siguiente.

````python
# Clase Vehiculo
from abc import ABC, abstractmethod

class Vehiculo(ABC):
    def __init__(self, marca, modelo, kilometraje=0):
        self._marca = marca
        self._modelo = modelo
        self._kilometraje = kilometraje

    def conducir(self, kilometros):
        self._kilometraje += kilometros

    # Falta incluir las propiedades

    @abstractmethod
    def info(self):
        pass
````

````python
# Clase coche
class Coche(Vehiculo):
    def __init__(self, marca, modelo, combustible='', kilometraje=0):
        super().__init__(marca, modelo, kilometraje)
        self._combustible = combustible

    def info(self):
        return f'{self._marca}, {self._modelo}, {self._kilometraje}, {self._combustible}'
    
    def esUsado(self):
        if self._kilometraje > 0:
            return True
        else:
            return False


# === TESTEAR COCHE ==== 

#cocheA = Coche("Ford", "Focus", "Gasolina")
#assert(cocheA.esUsado() is False)
#cocheA.conducir(1000)
#assert(cocheA.esUsado() is True)
#assert(cocheA.info() == "Ford Focus, Gasolina, 1000 km")

#cocheUsado = Coche("Citroen", "C4", "Diésel", 100000)
#assert(cocheUsado.esUsado() is True)
#assert(cocheUsado.info() == "Citroen C4, Diésel, 100000 km")
````
